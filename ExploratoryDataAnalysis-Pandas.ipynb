```python
import pandas as pd
```

```python
#reading the data from the file
df = pd.read_csv(r"/Users/shivakruthi/Downloads/dirty_cafe_sales.csv")
```

```python
#this calls the data in the file
df
```

```python
#removing duplicates from the data set and saving it
df = df.drop_duplicates()
df
```

```python
#trying to unnecessary values at the end of the data
```

```python
#replacing NaN values with empty value
df=df.fillna('')
df
```

```python
#replacing UNKNOWN values with empty value
df = df.replace('UNKNOWN', '')
df
```

```python
#replacing ERROR values with empty value
df = df.replace('ERROR', '')
df
```

```python
#replacing Nan values in the ITEM column with unknown value
df['Item'] = df['Item'].replace('','Unknown')
df
```

```python
#Converting values in Quantity column to Numeric/int
df['Quantity'] = pd.to_numeric(df['Quantity'])
df
```

```python
#filling missing values in quantity with median
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
df
```

```python
#Converting values in Price per Unit column to Numeric/int
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'])
df
```

```python
#filling missing values in Price with median
df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Price Per Unit'].median())
df
```

```python
#Converting values in Total Spent column to Numeric/int
df['Total Spent'] = pd.to_numeric(df['Total Spent'])
df
```

```python
#filling missing values in Total using formulae
#total= Quantity* Price Per Unit
#total
```

```python
df['Total Spent'] = df['Total Spent'].fillna(df['Total Spent'].median())
df
```

```python
#replacing empty values in the Payment method column with unknown value
df['Payment Method'] = df['Payment Method'].replace('','Unknown')
df
```

```python
#replacing empty values in the Location column with unknown value
df['Location'] = df['Location'].replace('','Unknown')
df
```

```python
#verifying if all the dates are in the same format
#for that first I am importing library
import pandas as pd
from datetime import datetime
```

```python
try:
    pd.to_datetime(df['Transaction Date'], format='%Y-%m-%d')
except ValueError:
    print("Dates are not in the same format.")
else:
    print("Dates are in the same format.")
```

```python
# Convert 'Transaction Date' column to datetime
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])
```

```python
#verifying if the Transaction Date Column has any null values
has_nulls = df['Transaction Date'].isnull().any()

if has_nulls:
    print("The 'Transaction Date' column contains null values.")
else:
    print("The 'Transaction Date' column does not contain null values.")
```

```python
# Filling any null values Transaction Date Column has.
df['Transaction Date'] = df['Transaction Date'].fillna(method='ffill') 
df
```
